## Correlation Analysis

Correlation analysis helps to:

- Identify strongly related features
- Detect multicollinearity
- Understand relationships with the target variable

```python
plt.figure(figsize=(12,10))
corr_matrix = df.corr()

sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.show()
```

## Feature vs Target Relationships

We examine how individual features relate to the degradation coefficient.

```python
for col in df.columns[:-1]:  # exclude target
    plt.figure(figsize=(6,4))
    plt.scatter(df[col], df[target_column], alpha=0.5)
    plt.xlabel(col)
    plt.ylabel(target_column)
    plt.title(f"{col} vs {target_column}")
    plt.show()
```

## Outlier Inspection

Boxplots are used to visually inspect extreme values that may influence model performance.

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot of Features")
plt.show()
```
